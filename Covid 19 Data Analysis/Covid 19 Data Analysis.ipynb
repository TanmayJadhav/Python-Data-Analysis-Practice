```python
import pandas as pd
```

```python
data = pd.read_csv('Covid 19 Data.csv')
```

```python
data
```

```python
data.count()
```

```python
data.isnull().sum()
```

## Show the number of confirmed,deaths and recovered in each regions

```python
data.groupby('Region')['Deaths','Confirmed','Recovered'].sum()

data.groupby('Region').sum()
```

## Remove all the records where confirmed cases are less than 10

```python
data = data[ ~ (data.Confirmed < 10)] 
```

```python
data
```

## In which region maximum number of confirmed cases where recorded

```python
data.groupby('Region').Confirmed.sum().sort_values(ascending=False)
```

## In which region minimum number of death cases were recorded

```python
data.groupby('Region').Deaths.min().sort_values(ascending=True)
```

## How many confirmed , death and recovered case were recorded from India till 29 April 2020

```python
data[data.Region == 'India']

# data[(data.Date == '4/29/2020') & (data.Region == 'India')]
```

## Sort the data wrt No. of confirmed cases in ascending order

```python
data.sort_values(by = ['Confirmed'],ascending=True)
```

## Sort the data in No. of Recovered by descending order

```python
data.sort_values('Recovered',ascending=False)
```
